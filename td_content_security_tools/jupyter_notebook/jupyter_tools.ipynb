# jupyter notebook 服务器配置与使用

## 1. jupyter 安装（服务端）

`pip install jupyter `# Anaconda 自带 jupyter 不需要安装

## 2. jupyter 配置 （服务端）

- 运行命令 `jupyter notebook --generate-config` 生成配置文件 `~/.jupyter/jupyter_notebook_config.py`
- 编辑配置文件加入以下内容：


```python
# --------------------------------------------------------------------------------------
c = get_config()
c.NotebookApp.allow_remote_access = True
c.NotebookApp.ip = "*"                              # 或者 c.NotebookApp.ip = "0.0.0.0"
c.NotebookAPp.open_browser = False                  # 不开启浏览器

# 访问密码，这个可以通过程序生成
c.NotebookApp.password = 'sha1:b7ca5bda60eb:0aa2376fa800aefb1680fa6c71781634182fa9c0'
c.NotebookApp.port= 8888                            # 服务端口号

c.NotebookApp.notebook_dir = "/home/user/workspace" # 工作目录
c.NotebookApp.terminals_enabled = True              # 允许开启终端
# --------------------------------------------------------------------------------------
```

密码生成代码 `generate_password.py`
```python
from notebook.auth import passwd
pwd = passwd('123456') # 假设密码是123456, 同样的密码每次输出的sha1都不一样的
print(pwd)
# sha1:b7ca5bda60eb:0aa2376fa800aefb1680fa6c71781634182fa9c0'\
```

- 启动服务：命令行运行 `jupyter notebook` 或者 `nohup jupyter notebook &`


## 3. 使用（客户端）

打开浏览器，使用服务器ip和端口号即可访问使用
